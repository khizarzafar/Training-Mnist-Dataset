```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
```

Divides the dataset into training and testing sets. x_train contains the images and y_train contains the labels. Similarly, x_test contains the images and y_test contains the corrresponding labels.

```python
(x_train,y_train) , (x_test,y_test) = mnist.load_data()
```

```python
print(f"The training dataset shape is {x_train.shape}")
```

```python
print(f"The testing dataset shape is {x_test.shape}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,2))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(x_train[i],cmap="grey")
    plt.axis("off")
    plt.title(str(y_train[i]))
    plt.tight_layout()
```
